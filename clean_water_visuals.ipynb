```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
all_yearly = pd.read_csv('/content/part-000y-at-least-basic.csv')
all_yearly.columns = ['year', 'Percent with clean water']
countries = pd.read_csv('/content/part-sdg-basic.csv')
countries.columns = ['regions', 'Percent with clean water']
all_countries_yearly = pd.read_csv('/content/part-sdg-yearly.csv')
all_countries_yearly.columns = ['regions', 'year', 'Percent with clean water']
```

```python
all_yearly
```

```python
all_yearly_line = sns.lineplot(data=all_yearly, x = 'year', y = 'Percent with clean water')
all_yearly_line.set_title('Percent of population with access to clean water')
all_yearly_line.set_xlabel('Year')
all_yearly_line.set_ylabel('Percent of population')
plt.show()
```

```python
countries_sorted = countries.sort_values('Percent with clean water', ascending=False)
bar_plot = sns.barplot(data=countries_sorted, x = 'Percent with clean water', y = 'regions', order=countries_sorted['regions'])
bar_plot.set_title('Percent of population with access to clean water')
bar_plot.set_xlabel('Percent of population')
bar_plot.set_ylabel('Regions')
plt.show()
```

```python
all_countries_yearly
lineplot = sns.lineplot(data=all_countries_yearly, x = 'year', y = 'Percent with clean water', hue = 'regions')
lineplot.set_title('Percent of population with access to clean water')
lineplot.set_xlabel('Year')
lineplot.set_ylabel('Percent of population')
lineplot.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
